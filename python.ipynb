# mathplotlib:-


```python
import matplotlib.pyplot as plt
x=np.arange(0,100,0.2)
y=np.cos(x)
fig=plt.figure(figsize=(10,10))
plt.scatter(x,y)
plt.show()
```

# pie chat:-

```python
students={"Ex":28,"A":10,"B":26,"C":12,"D":10}
categories=list(students.keys())
percentages=list(students.values())
```

```python
fig=plt.figure(figsize=(6,6))
plt.pie(x=percentages,labels=categories,explode=[0,0,0,0.5,0],shadow=True)
plt.legend(["Ex 90-100%","A 80-90%","B 70-80%","C 60-70%","50-60%"])
plt.show()
```

# bar graph:-

```python
plt.bar(x=categories,height=percentages)
fig=plt.figure9figsize=(6,6)
plt.xlabel("categories")
plt.ylabel("percentages")
plt.show()
```

```python
plt.barh(y=categories,width=percentages)
fig=plt.figure9figsize=(6,6)
plt.xlabel("categories")
plt.ylabel("percentages")
plt.show()
```

# histogram

```python
marks=[90,69,80,66,86]
print(len(marks))
fig=plt.figure(figsize=(6,6))
plt.hist(marks)
plt.show()
```
